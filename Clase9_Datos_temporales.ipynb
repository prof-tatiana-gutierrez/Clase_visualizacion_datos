```python
sns.set(font_scale=1.4)

def reformat_large_tick_values(tick_val, pos):
    """
    Turns large tick values (in the billions, millions and thousands) such as 4500 into 4.5K and also appropriately turns 4000 into 4K (no zero after the decimal).
    """
    if tick_val >= 1000000000:
        val = round(tick_val/1000000000, 1)
        new_tick_format = '{:}B'.format(val)
    elif tick_val >= 1000000:
        val = round(tick_val/1000000, 1)
        new_tick_format = '{:}M'.format(val)
    elif tick_val >= 1000:
        val = round(tick_val/1000, 1)
        new_tick_format = '{:}K'.format(val)
    elif tick_val < 1000:
        new_tick_format = round(tick_val, 1)
    else:
        new_tick_format = tick_val

    # make new_tick_format into a string value
    new_tick_format = str(new_tick_format)
    
    # code below will keep 4.5M as is but change values such as 4.0M to 4M since that zero after the decimal isn't needed
    index_of_decimal = new_tick_format.find(".")
    
    if index_of_decimal != -1:
        value_after_decimal = new_tick_format[index_of_decimal+1]
        if value_after_decimal == "0":
            # remove the 0 after the decimal point since it's not needed
            new_tick_format = new_tick_format[0:index_of_decimal] + new_tick_format[index_of_decimal+2:]
            
    return new_tick_format

#https://dfrieds.com/data-visualizations/how-format-large-tick-values.html
```

```python
import matplotlib.ticker as tick
sns.lineplot(data=Conteo_leticia, x="Fecha de Publicacion del Proceso", y="Valor")
plt.ticklabel_format(style='plain', axis='y')
ax = plt.gca()
ax.yaxis.set_major_formatter(tick.FuncFormatter(reformat_large_tick_values))
plt.show()
```

# Gráfico área radial

https://www.python-graph-gallery.com/391-radar-chart-with-several-individuals

```python
from math import pi
# Datos
df = pd.DataFrame({
'group': ['2017','2018','2019','2020'],
'Enero': [38, 1.5, 30, 4],
'Febrero': [29, 10, 9, 34],
'Marzo': [8, 39, 23, 24],
'Abril': [7, 31, 33, 14],
'Mayo': [28, 15, 32, 14],
'Junio': [28, 17, 32, 14],
'Julio': [38, 5, 30, 4],
'Agosto': [29, 8, 9, 34],
'Septiembre': [8, 34, 23, 24],
'Octubre': [7, 70, 33, 14],
'Noviembre': [28, 67, 32, 14],
'Diciembre': [28, 15, 32, 14], 
})
df
```

```python
# ------- PARTE 1: Crear el backgroud
 
# Número de variables numéricas
categories=list(df)[1:]
N = len(categories)
print(categories)
print(N)
```

```python
maxnum = df[categories].values.max()
maxnum
```

```python
# ¿Cuál será el ángulo de cada eje en el gráfico? (dividimos la parcela / número de variable)
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]
 
# Iniciar el gráfico de area radial (alias spider plot)
ax = plt.subplot(111, polar=True)
 
# Si tu quieres que el primer eje quede en el top
ax.set_theta_offset(pi / 2)
ax.set_theta_direction(-1)

# Dibujar un por uno el eje y agregar labels
plt.xticks(angles[:-1], categories)
 
# Dibujar los labels de y (ojo con el min y max)
ax.set_rlabel_position(0)
plt.ylim(0,maxnum)
```

```python
# ¿Cuál será el ángulo de cada eje en el gráfico? (dividimos la parcela / número de variable)
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]
 
# Iniciar el gráfico de area radial (alias spider plot)
ax = plt.subplot(111, polar=True)
 
# Si tu quieres que el primer eje quede en el top
ax.set_theta_offset(pi / 2)
ax.set_theta_direction(-1)

# Dibujar un por uno el eje y agregar labels
plt.xticks(angles[:-1], categories)
 
# Dibujar los labels de y (ojo con el min y max)
ax.set_rlabel_position(0)
plt.ylim(0,maxnum)

# ------- PARTE 2: Agregar gráficos (uno por uno :( )
 
# Año 2017
values=df.loc[0].drop('group').values.flatten().tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label="2017")
ax.fill(angles, values, 'blue', alpha=0.1)
 
# Año 2018
values=df.loc[1].drop('group').values.flatten().tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label="2018")
ax.fill(angles, values, 'blue', alpha=0.1)

# Año 2019
values=df.loc[2].drop('group').values.flatten().tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label="2019")
ax.fill(angles, values, 'blue', alpha=0.1)
 
# Año 2020
values=df.loc[3].drop('group').values.flatten().tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label="2020")
ax.fill(angles, values, 'blue', alpha=0.1)
 
# Agregar leyenda
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

# TARAAAAN
plt.show()
```
